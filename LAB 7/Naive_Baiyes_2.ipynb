```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
```

```python

#Step 1: Create the dataset
data={
    'Weather':['sunny','sunny','overcast','rain','rain'],
    'Wind':['weak','strong','weak','weak','strong'],
    'Play':['Yes','No','Yes','Yes','No']
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
#Step 2: Encode categorical variables
le_weather=LabelEncoder()
le_wind=LabelEncoder()
le_play=LabelEncoder()
```

```python
df['Weather']=le_weather.fit_transform(df['Weather'])
df['Wind']=le_wind.fit_transform(df['Wind'])
df['Play']=le_play.fit_transform(df['Play'])
```

```python
#Step 3:Features (X) and Target (y)
X=df[['Weather','Wind']]
y=df['Play']
```

```python
#Step 4: Train Naive Bayes
model=CategoricalNB()
model.fit(X,y)
```

```python
#Step 5: Predict on the same data
predictions=model.predict(X)
```

```python
#Step 6:Show results with decoded labels
df['Predicted_Play']=le_play.inverse_transform(predictions)
df['Actual_Play']=le_play.inverse_transform(y)
```
