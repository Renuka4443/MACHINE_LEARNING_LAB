```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LogisticRegression 
```

```python
# Sample data: hours studied
X = np.array([[1],[2],[3],[4],[5],[6],[7],[8]])

# Labels: 0 = Fail, 1 = Pass
y = np.array([0,0,0,0,1,1,1,1])
```

```python
# Train logistic regression model 
model = LogisticRegression()
model.fit(X, y)
```

```python
# Generate range of study hours for smooth curve
X_test = np.linspace(0, 9, 100).reshape(-1, 1)

# Predict probabilities for positive class (Pass)
y_prob = model.predict_proba(X_test)[:, 1]  # Take probability of class 1 only
```

```python
# Plot data points
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='black', label="Data points (Fail=0, Pass=1)")

# Plot sigmoid curve (probability of passing)
plt.plot(X_test, y_prob, color='blue', label="Logistic Regression curve")

# Add decision threshold at 0.5
plt.axhline(0.5, color='red', linestyle='--', label="Threshold = 0.5")

# Add labels and title
plt.title("Logistic Regression: Study Hours vs. Pass/Fail")
plt.xlabel("Study Hours")
plt.ylabel("Probability of Passing")

# Add legend and grid
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
